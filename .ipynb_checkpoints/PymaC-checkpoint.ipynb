```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import sem
import statistics

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_drug_data_df = pd.read_csv(mouse_drug_data_to_load)
cl_trial_data_df = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
combined_mouse_data = pd.merge(cl_trial_data_df, mouse_drug_data_df, on="Mouse ID", how="left")

# Display the data table for preview
combined_mouse_data.head(5)
```

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
tumor_vol = combined_mouse_data.groupby(["Drug", "Timepoint"])["Tumor Volume (mm3)"].mean().reset_index()

# Convert to DataFrame
type(tumor_vol)

# Preview DataFrame
tumor_vol.head(8)
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
drug_name = combined_mouse_data[["Drug", "Timepoint", "Tumor Volume (mm3)"]]
drug_name = drug_name.sort_values(by = ["Drug", "Timepoint"])

drug_name['Vol Mean'] = drug_name.groupby(["Drug", "Timepoint"])["Tumor Volume (mm3)"].transform(pd.Series.mean)
drug_name['Difference'] = (drug_name["Tumor Volume (mm3)"] - drug_name["Vol Mean"])
drug_name['Differencesq'] = drug_name['Difference']**2
# drug_name

drug_name_pivot = pd.pivot_table(drug_name,index=["Drug","Timepoint"],values=["Differencesq"], aggfunc=[np.sum, len]).reset_index()
# print (drug_name_pivot.columns.tolist())
sumDiffsq = drug_name_pivot[('sum', 'Differencesq')]
lenDiffsq = (drug_name_pivot[('len', 'Differencesq')])
lenDiffsqmone = (drug_name_pivot[('len', 'Differencesq')] - 1)
drug_name_pivot["Variance"] = (sumDiffsq/lenDiffsqmone)
drug_name_pivot["Stand Dev"] = drug_name_pivot["Variance"]**(.5)
drug_name_pivot["Stand Error"] = drug_name_pivot["Stand Dev"]/(lenDiffsq**(.5))
drug_name_pivot[["Drug", "Timepoint", "Stand Error"]].head(8)

# Convert to DataFrame

# Preview DataFrame
```

```python
# Minor Data Munging to Re-Format the Data Frames

# Preview that Reformatting worked

combined_mouse_data.pivot_table(index='Timepoint', columns='Drug', values=['Tumor Volume (mm3)'])
```

```python
means = list(tumor_vol["Tumor Volume (mm3)"])
meanCapomulin = means[0:10]
meanInfubinol = means[20:30]
meanKetapril = means[30:40]
meanPlacebo = means[50:60]

sems = list(drug_name_pivot["Stand Error"])
semCapomulin = sems[0:10]
semInfubinol = sems[20:30]
semKetapril = sems[30:40]
semPlacebo = sems[50:60]

xname = list(drug_name_pivot["Timepoint"])
xname = xname[0:10]
# print(xname)

drugnames = drug_name_pivot["Drug"].unique()
```

```python
fig, ax = plt.subplots()

ax.errorbar(xname, meanCapomulin, yerr=semCapomulin, fmt="-", marker="o", color="brown", alpha=0.75, label="Capomulin")
ax.errorbar(xname, meanInfubinol, yerr=semInfubinol, fmt="-", marker="*", color="red", alpha=0.75, label="Infubinol")
ax.errorbar(xname, meanKetapril, yerr=semKetapril, fmt="-", marker="^", color="green", alpha=0.75, label="Ketapril")
ax.errorbar(xname, meanPlacebo, yerr=semPlacebo, fmt="-", color="blue", alpha=0.75, label="Placebo")

ax.set_xlabel("Time (Days)")
ax.set_ylabel("Tumor Volume (mm3)")

plt.title("Tumor Response to Treatment")
plt.legend(loc="upper left")
plt.grid()
plt.show()
```

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
met_site = combined_mouse_data.groupby(["Drug", "Timepoint"])["Metastatic Sites"].mean().reset_index()

# Convert to DataFrame
type(met_site)

# Preview DataFrame
met_site.head(8)
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
met_site_data = combined_mouse_data[["Drug", "Timepoint", "Metastatic Sites"]]
met_site_data = met_site_data.sort_values(by = ["Drug", "Timepoint"])

met_site_data['Vol Mean'] = met_site_data.groupby(["Drug", "Timepoint"])["Metastatic Sites"].transform(pd.Series.mean)
met_site_data['Difference'] = (met_site_data["Metastatic Sites"] - met_site_data["Vol Mean"])
met_site_data['Differencesq'] = met_site_data['Difference']**2


met_site_pivot = pd.pivot_table(met_site_data, index=["Drug","Timepoint"],values=["Differencesq"], aggfunc=[np.sum, len]).reset_index()

ms_sumDiffsq = met_site_pivot[('sum', 'Differencesq')]
ms_lenDiffsq = (met_site_pivot[('len', 'Differencesq')])
ms_lenDiffsqmone = (met_site_pivot[('len', 'Differencesq')] - 1)
met_site_pivot["Variance"] = (ms_sumDiffsq/ms_lenDiffsqmone)
met_site_pivot["Stand Dev"] = met_site_pivot["Variance"]**(.5)
met_site_pivot["Stand Error"] = met_site_pivot["Stand Dev"]/(ms_lenDiffsq**(.5))
met_site_pivot[["Drug", "Timepoint", "Stand Error"]].head(8)
```

```python
mousecount_data = combined_mouse_data[["Drug", "Timepoint", "Mouse ID"]]
mousecount_data = mousecount_data.sort_values(by = ["Drug", "Timepoint"])

mousecount_data['Mouse Count'] = mousecount_data.groupby(["Drug", "Timepoint"])["Mouse ID"].transform(pd.Series.nunique)
mousecount_data['Max Count'] = mousecount_data.groupby(['Drug'])['Mouse Count'].transform(max)
mousecount_pivot = pd.pivot_table(mousecount_data, index=["Drug","Timepoint"],values=["Max Count", "Mouse Count"], aggfunc=[np.mean]).reset_index()

maxcountvar = mousecount_pivot[('mean', 'Max Count')]
mousecountvar = mousecount_pivot[('mean', 'Mouse Count')]

mousecount_pivot["% Decline"] = mousecountvar/maxcountvar
```

```python
mc_percent = list(mousecount_pivot["% Decline"])
mc_percentCapomulin = mc_percent[0:10]
mc_percentInfubinol = mc_percent[20:30]
mc_percentKetapril = mc_percent[30:40]
mc_percentPlacebol = mc_percent[50:60]

mc_xname = list(mousecount_pivot["Timepoint"])
mc_xname = mc_xname[0:10]
```

```python
Capomulin, = plt.plot(mc_xname, mc_percentCapomulin, marker="^", color='purple', label='Capomulin')
Infubinol, = plt.plot(mc_xname, mc_percentInfubinol, marker="o", color='yellow', label='Infubinol')
Ketapril, = plt.plot(mc_xname, mc_percentKetapril, marker="*", color='pink', label='Ketapril')
Placebo, = plt.plot(mc_xname, mc_percentPlacebol, color='green', label='Placebo')

plt.title("Survival During Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Survival Rate (%)")

plt.legend(handles=[Capomulin, Infubinol, Ketapril, Placebo], loc="lower left")
plt.grid()
plt.show()
```

```python
 # Calculate the percent changes for each drug
com_query = combined_mouse_data.pivot_table(index='Timepoint', columns='Drug', values=['Tumor Volume (mm3)'])
percentq = com_query.query('Timepoint == [0, 45]')
percent_change = np.transpose(percentq)
# print (percent_change.columns.tolist())

percent_change["% Change"] = ((percent_change[45]/percent_change[0] - 1)*100)
percent_change['% Change']
```

```python
# Store all Relevant Percent Changes into a Tuple
percent_change = percent_change.reindex(percent_change['% Change'].sort_values(ascending=True).index)
drug_name_order = percent_change.index.tolist()
# print(drug_name_order)

drug_name_arr = np.array(drug_name_order)
drug_name_arr = drug_name_arr.T[1]
# print(drug_name_arr)

percent_change_list = list(percent_change["% Change"])
percent_change_tuple = tuple(percent_change_list)


# Splice the data between passing and failing drugs
neg_tuple = percent_change_tuple[0:2]
pos_tuple = percent_change_tuple[2:10]


# Orient widths. Add labels, tick marks, etc. 
xval = drug_name_arr
yval = percent_change_tuple
x_axis = np.arange(len(percent_change_tuple))

bar_graph = plt.figure(figsize=(15,5))
bar_graph = plt.bar(x_axis, yval, color="red", align="center")

for i, v in enumerate(yval):
    plt.text(i, v/yval[i], '{0:.0%}'.format((yval[i]/100)), fontsize=10)
    
bar_graph[0].set_color('green')
bar_graph[1].set_color('green')
bar_graph


tick_locations = [value + .5 for value in x_axis]
plt.xticks(tick_locations, xval)
plt.xlim(-0.75, len(x_axis)-0.25)

plt.title("Tumor Change Over 45 Day Treatment")
plt.ylabel("% Tumor Volume Change")

# Use functions to label the percentages of changes
# Call functions to implement the function calls

# Save the Figure
plt.savefig("./PymaBarGraph.png")

# Show the Figure
plt.grid(True)
fig.show()
```
