```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import sem
import statistics

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_drug_data_df = pd.read_csv(mouse_drug_data_to_load)
cl_trial_data_df = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
combined_mouse_data = pd.merge(cl_trial_data_df, mouse_drug_data_df, on="Mouse ID", how="left")

# Display the data table for preview
combined_mouse_data.head(5)
```

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
tumor_vol = combined_mouse_data.groupby(["Drug", "Timepoint"])["Tumor Volume (mm3)"].mean().reset_index()

# Convert to DataFrame
type(tumor_vol)

# Preview DataFrame
tumor_vol.head(8)
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
drug_name = combined_mouse_data[["Drug", "Timepoint", "Tumor Volume (mm3)"]]
drug_name = drug_name.sort_values(by = ["Drug", "Timepoint"])

drug_name['Vol Mean'] = drug_name.groupby(["Drug", "Timepoint"])["Tumor Volume (mm3)"].transform(pd.Series.mean)
drug_name['Difference'] = (drug_name["Tumor Volume (mm3)"] - drug_name["Vol Mean"])
drug_name['Differencesq'] = drug_name['Difference']**2
# drug_name

drug_name_pivot = pd.pivot_table(drug_name,index=["Drug","Timepoint"],values=["Differencesq"], aggfunc=[np.sum, len]).reset_index()
# print (drug_name_pivot.columns.tolist())
sumDiffsq = drug_name_pivot[('sum', 'Differencesq')]
lenDiffsq = (drug_name_pivot[('len', 'Differencesq')])
lenDiffsqmone = (drug_name_pivot[('len', 'Differencesq')] - 1)
drug_name_pivot["Variance"] = (sumDiffsq/lenDiffsqmone)
drug_name_pivot["Stand Dev"] = drug_name_pivot["Variance"]**(.5)
drug_name_pivot["Stand Error"] = drug_name_pivot["Stand Dev"]/(lenDiffsq**(.5))
drug_name_pivot[["Drug", "Timepoint", "Stand Error"]].head(8)

# Convert to DataFrame

# Preview DataFrame
```

```python
means = list(tumor_vol["Tumor Volume (mm3)"])
sems = list(drug_name_pivot["Stand Error"])
xname = list(drug_name_pivot["Timepoint"])
drugnames = drug_name_pivot["Drug"].unique()
```

```python
fig, ax = plt.subplots()

ax.errorbar(xname, means, yerr=sems, fmt="-", color="b", alpha=0.5, label=drugnames)

ax.set_xlim(0, 50) 

ax.set_xlabel("Time (Days)")
ax.set_ylabel("Tumor Volume (mm3)")


plt.grid()
plt.show()
```

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
met_site = combined_mouse_data.groupby(["Drug", "Timepoint"])["Metastatic Sites"].mean().reset_index()

# Convert to DataFrame
type(met_site)

# Preview DataFrame
met_site.head(8)
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
met_site_data = combined_mouse_data[["Drug", "Timepoint", "Metastatic Sites"]]
met_site_data = met_site_data.sort_values(by = ["Drug", "Timepoint"])

met_site_data['Vol Mean'] = met_site_data.groupby(["Drug", "Timepoint"])["Metastatic Sites"].transform(pd.Series.mean)
met_site_data['Difference'] = (met_site_data["Metastatic Sites"] - met_site_data["Vol Mean"])
met_site_data['Differencesq'] = met_site_data['Difference']**2


met_site_pivot = pd.pivot_table(met_site_data, index=["Drug","Timepoint"],values=["Differencesq"], aggfunc=[np.sum, len]).reset_index()

ms_sumDiffsq = met_site_pivot[('sum', 'Differencesq')]
ms_lenDiffsq = (met_site_pivot[('len', 'Differencesq')])
ms_lenDiffsqmone = (met_site_pivot[('len', 'Differencesq')] - 1)
met_site_pivot["Variance"] = (ms_sumDiffsq/ms_lenDiffsqmone)
met_site_pivot["Stand Dev"] = met_site_pivot["Variance"]**(.5)
met_site_pivot["Stand Error"] = met_site_pivot["Stand Dev"]/(ms_lenDiffsq**(.5))
met_site_pivot[["Drug", "Timepoint", "Stand Error"]].head(8)
```

```python
mousecount_data = combined_mouse_data[["Drug", "Timepoint", "Mouse ID"]]
mousecount_data = mousecount_data.sort_values(by = ["Drug", "Timepoint"])

mousecount_data['Mouse Count'] = mousecount_data.groupby(["Drug", "Timepoint"])["Mouse ID"].transform(pd.Series.nunique)

mousecount_data['Max Count'] = mousecount_data.groupby(['Drug'])['Mouse Count'].transform(max)

mousecount_pivot = pd.pivot_table(mousecount_data, index=["Drug","Timepoint"],values=["Max Count", "Mouse Count"], aggfunc=[np.mean]).reset_index()
# print (mousecount_pivot.columns.tolist())

maxcountvar = mousecount_pivot[('mean', 'Max Count')]
mousecountvar = mousecount_pivot[('mean', 'Mouse Count')]

mousecount_pivot["% Decline"] = mousecountvar/maxcountvar
mousecount_pivot[["Drug", "Timepoint", "% Decline"]].head(8)
```
